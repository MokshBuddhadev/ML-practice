```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])           
y_train = np.array([300.0, 500.0]) 
```

```python
print(x_train)
print(y_train)
```

```python
def compute_CostFunction(x,y,w,b):
    m=len(x)
    f_wb=np.zeros(0)
    cost_sum=0
    for i in range (m):
        f_wb=w*x[i]+b
        cost=(f_wb-y[i])**2
        cost_sum=cost+cost_sum
    total_CostFunction=(1/(2*m))*cost_sum
    return total_CostFunction
        
```

```python
cost_function=compute_CostFunction(x_train,y_train,50,100)
```
