we are going to create a vector rather than creating new variables for each and every region and we will also define weights and create a function
the function will provide realtion between the factors and weights therby providing information about the yield

```python
kanto=[73,67,43]
johto=[91,88,64]
```

```python
w1=0.3
w2=0.2
w3=0.5
```

```python
weights=[w1,w2,w3]
```

```python
def calc_yield(region,weights):
    result=0
    for x,w in zip(region,weights):
        result+=x*w
    return result
```

```python
calc_yield(kanto,weights)
```
