```python
import numpy as np
```

```python
climate_data=np.genfromtxt("climate.txt",delimiter=",",skip_header=1)
```

```python
climate_data
```

```python
climate_data.shape
```

```python
w1,w2,w3=0.3,0.2,0.5
```

```python
weights=w1,w2,w3
```

```python
weights
```

```python
yield_apples = climate_data @ weights
```

```python
yield_apples
```

```python
climate_results=np.concatenate((climate_data,yield_apples.reshape(10000,1)),axis=1)
```

```python
climate_results
```

```python
np.savetxt('climate_results.txt',
           climate_results,
           fmt='%.2f',
           header='temperature,rainfall,humidity,yield_of_apples',
           comments=' ')
```
