```python
# Convert the 'date' column to datetime format
covid_df['date'] = pd.to_datetime(covid_df['date'])
```

```python
covid_df['date']
```

```python
# Convert date column to datetime and extract year
covid_df['year'] = pd.to_datetime(covid_df['date']).dt.year
```

```python
covid_df['month'] = pd.to_datetime(covid_df['date']).dt.month
```

```python
covid_df['month']
```

```python
# here we are grouping the rows in terms of the column month and then we are taking the columns of cases and tests and grouping them in terms of sum in terms of months
```

```python
monthly_group = covid_df.groupby('month')
```

```python
monthly_group[['number_of_cases','number_of_tests']].sum()
```

```python
monthly_group[['number_of_cases','number_of_tests']].mean()
```
