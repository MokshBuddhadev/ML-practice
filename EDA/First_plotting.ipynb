```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
years = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]

yield_apples = [120, 135, 128, 140, 150, 145, 160, 155, 170]
yield_oranges = [100, 110, 108, 115, 120, 118, 130, 125, 135]
```

```python
plt.plot(yield_apples)
plt.show()
```

```python
plt.plot(years,yield_apples)
plt.show()
```

```python
plt.plot(years,yield_apples)
plt.plot(years,yield_oranges)

plt.xlabel('years')
plt.ylabel('yield in tons')

plt.title('Yield of Apples and Oranges')
plt.legend(['Apples','Oranges'])

plt.show()
```

```python
plt.plot(years, yield_apples, "g--o")   # green, dashed, circle
plt.plot(years, yield_oranges, "b-x") # blue, solid , x marker
plt.show()
```

```python
plt.figure(figsize=(11,6)) # this statement should be at the start 
plt.plot(years, yield_apples, "g--o")   # green, dashed, circle
plt.plot(years, yield_oranges, "b-x") # blue, solid , x marker
plt.show()
```

```python
# scatter plots
```

```python
flower_df=sns.load_dataset('iris')
```

```python
flower_df
```

```python
flower_df.species.unique() #to find out unique values of species of the flowers
```

```python
plt.plot(flower_df.sepal_length,flower_df.sepal_width)
plt.show()
```

```python
# the above graph shows that plotting a line chart for the above data is not helpful so we need an alternative
```

```python
sns.scatterplot(x=flower_df.sepal_length,y=flower_df.sepal_width)
plt.show()
```

```python
sns.scatterplot(x=flower_df.sepal_length,y=flower_df.sepal_width,hue=flower_df.species)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('Species of Flowers')
sns.scatterplot(x=flower_df.sepal_length,y=flower_df.sepal_width,hue=flower_df.species)
plt.show()
```

```python
sns.scatterplot(x='sepal_length',
                y='sepal_width',
                hue='species',
                data=flower_df)
plt.show()
```

```python
# Bar Plots
```

```python
bills_df=sns.load_dataset('tips')
```

```python
bills_df
```

```python
grouped_df = bills_df.groupby('day')[['total_bill']].mean()
grouped_df
```

```python
# here we are using index because in the grouped_df the day column has become an index and it does not exist as in independent column
plt.bar(grouped_df.index,grouped_df.total_bill)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=grouped_df)
plt.show()
```

```python
sns.barplot(y='day',x='total_bill',data=grouped_df)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=bills_df)
plt.show()
```

```python
df=sns.load_dataset('flights')
```

```python
df
```

```python
flights_df=sns.load_dataset('flights').pivot(index='month',columns='year',values='passengers')
```

```python
flights_df
```

```python
sns.heatmap(flights_df)
plt.show()
```

```python
sns.heatmap(
    flights_df,
    annot=True,
    fmt="d",
    cmap="Blues",
    linewidths=0.5,
    cbar_kws={'label': 'Passengers'}
)
plt.show()
```
