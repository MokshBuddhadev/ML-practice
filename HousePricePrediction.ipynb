```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([2.50,5.37,7.79,10.78,12.50,15.56,17.97,20.01])
y_train=np.array([175,256,315,458,568,675,796,897])
```

```python
print(f"The sizes of the house in sq.feet are: {x_train}")
print(f"The costs of the house in thousand dollars are: {y_train}")
```

```python
plt.scatter(x_train,y_train,marker='x',c='r')
plt.show()
```

```python
w_trial=50
b_trial=100
```

```python
def compute_function(x,w,b):
    m=len(x)
    f_wb_trial=np.zeros(m)
    for i in range(m):
        f_wb_trial[i]=w*x[i]+b
    return f_wb_trial
```

```python
output_trial=compute_function(x_train,w_trial,b_trial)
```

```python
print(output_trial)
```

```python
plt.plot(x_train,output_trial,c='b',label='prediction')
plt.scatter(x_train,y_train,marker='x',c='r',label='actual')
plt.legend()
plt.show()
```
