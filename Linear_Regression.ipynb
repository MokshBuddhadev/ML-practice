```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1.0,2.0])
y_train=np.array([300.0,500.0])
```

```python
print(f"the size of the house is {x_train[0]} and the price of the house is {y_train[0]}")
```

```python
m=len(x_train)
print(f"the number of training examples are: {m}")
```

```python
i=0
x_i=x_train[i]
y_i=y_train[i]
print(f"the selected training example is {x_i},{y_i}")
```

```python
plt.scatter(x_train,y_train,marker='x',c='r')
plt.title("Plotting the points")
plt.xlabel("size of the house in sq.feet")
plt.ylabel("price of the house in thousands")
plt.show()
```

```python
w=100
b=100
```

Below we are creating a function where the parameters are x, w and b and it computes the prediction based on the function f=wx+b


```python
def compute_output(x,w,b):
    m=len(x)
    f_x=np.zeros(m)
    for i in range(m):
        f_x[i]=w*x[i]+b
    return f_x
    
```

Here we are storing the values of prediction in the variable and pass in the numpy array x_train as the input

```python
output = compute_output(x_train,w,b,)
```

```python
print(output)
```

```python
plt.plot(x_train,output,c='r',label='prediction')
plt.scatter(x_train,y_train,marker='x',c='b',label='actual values')
plt.legend()
plt.show()
```

With the above plotting we can conclude that the selected values for the Weight(w) and Bias(b) does not fit the data well
